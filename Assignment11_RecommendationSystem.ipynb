# Preprocessing

```python
df.duplicated().value_counts()
```

```python
df.drop_duplicates(inplace=True)
df
```

# User Item Matrix

```python
userItem = df.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating')
userItem
```

```python
userItem=userItem.fillna(value=0.0)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
sim=1-pairwise_distances(userItem.values,metric='cosine')
sim
```

```python
np.fill_diagonal(sim,0)
```

```python
user_sim = pd.DataFrame(sim)
```

```python
# user-user matrix
user_sim
```

```python
user_sim.index = userItem.index
user_sim.columns = userItem.index
```

```python
#user - user matrix
user_sim
```

# Most similiar user pairs

```python
user_sim.idxmax(axis=1)
```

```python
# TOP 5 for each user id which have a similiar pair 
count = 1
for id in user_sim.index[1:]:
    df_new = pd.DataFrame(user_sim.iloc[count,0:])
    l=[]
    for i in list(range(0,5)):
        l.append(df_new.idxmax()[id])
        df_new.drop(index=df_new.idxmax()[id],axis=0,inplace=True)
    count = count+1
    if l!=[8, 9, 10, 12, 14]:
        print("user id:"+str(id)+" "*10+str(l))
    
```
